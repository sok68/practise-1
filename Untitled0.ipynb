```python

import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
```

```python
p_df=pd.read_csv('/content/fish_catch.csv')
p_df.head()
```

```python
imputer = SimpleImputer(strategy='most_frequent')
p_df['Sex'] = imputer.fit_transform(p_df[['Sex']])
p_df.head()
```

```python
label_encoder = LabelEncoder()
p_df['Species'] = label_encoder.fit_transform(p_df['Species'])
p_df.head()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(p_df['Species'], p_df['Weight'], c=p_df['Length1'], cmap='viridis', s=100, alpha=0.7)
plt.colorbar(label='Length1')
plt.xlabel('Species')
plt.ylabel('Weight')
plt.title('Зависимость веса рыбы от вида и длины 1')
plt.xticks(p_df['Species'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(p_df['Weight'], p_df['Length3'], c='blue', s=100, alpha=0.7)
plt.xlabel('Вес')
plt.ylabel('Длина3')
plt.title('Зависимость длины3 от веса')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(p_df['Sex'], p_df['Weight'], marker='o', color='b')
plt.xticks([0, 1], ['Женский', 'Мужской'])
plt.xlabel('Пол')
plt.ylabel('Вес')
plt.title('Зависимость веса рыбы от пола')
plt.grid(True)
plt.show()
```

```python
p_df=p_df.drop("Sex", axis=1)
X = p_df.drop("Weight", axis=1)
Y = p_df[["Weight"]]
Y.head()
```

```python

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

x_train.shape
```

```python
imputer = SimpleImputer(strategy='mean')

x_train = imputer.fit_transform(x_train)
y_train = imputer.fit_transform(y_train)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'R-squared (R^2): {r2:.2f}')
```

```python
import statsmodels.api as sm

X_train = sm.add_constant(x_train)
model_4 = sm.OLS(y_train, X_train).fit()
print(model_4.summary())
```

```python
p_df['Predicted_Weight'] = model.predict(X)
```

```python
p_df.head()
```

```python
plt.scatter(p_df['Weight'], p_df['Predicted_Weight'])
plt.xlabel('Фактический вес')
plt.ylabel('Предсказанный вес')
plt.title('Линейная зависимость')
plt.show()

```

```python

residuals = p_df['Weight'] - p_df['Predicted_Weight']
plt.scatter(p_df['Predicted_Weight'], residuals)
plt.xlabel('Предсказанный вес')
plt.ylabel('Остатки')
plt.title('Постоянная дисперсия')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

```python
from statsmodels.graphics.gofplots import qqplot
qqplot(residuals, line='s')
plt.title('QQ-график остатков')
plt.show()
```
